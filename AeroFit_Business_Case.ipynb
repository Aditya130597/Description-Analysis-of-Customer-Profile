##**AeroFit Business Case Study Using Python**

Aerofit is a leading brand in the field of fitness equipment.

AeroFit has continued to provide quality services that exceeds the expectations of their esteemed customers.

Aerofit provides a product range including machines such as treadmills, exercise bikes, gym equipment, and fitness accessories to cater to the needs of all categories of people.


**Data Set:**

● Product: Product Purchased KP281, KP481, or KP781

● Age: In years

● Gender: Male/Female

● Education: in years

● MaritalStatus: single or partnered

● Usage: average number of times the customer plans to use the treadmill each week

● Income: annual income (in $)

● Fitness: self-rated fitness on a 1-to-5 scale, where 1 is poor shape and 5 is the excellent shape.

● Miles: average number of miles the customer expects to walk/run each week

```python
#Importing Python Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Importing the Data Set
df=pd.read_csv('aerofit_treadmill.csv')
df_main=pd.read_csv('aerofit_treadmill.csv')
```

**Importing the dataset and performing usual data analysis steps like checking the structure & characteristics of the dataset**

```python
#Checking the data type of all columns in the given data set
df.info()
```

```python
#Familiarizing with the different rows and columns of the table
df
```

```python
df_product_price = pd.DataFrame({
    "Product":["KP281","KP481","KP781"],
    "Product_price":[1500,1750,2500]
})
df_product_price
```

```python
df = df.merge(df_product_price,on="Product",how = "left")
df
```

Categorising the age column intodifferent age groups
- Teen (0-21)
- Adult (22-35)
- Mid_age (36-45)
- Towards_old_age (>46)

```python
df["Age_category"] = df["Age"]
```

```python
df["Age_category"] = pd.cut(df["Age_category"], bins =[0,21,35,45,60], include_lowest=True, labels=["Teen","Adult","mid_age","Towards_old_age"])
df["Age_category"] =df["Age_category"].astype("object")
```

Clipping the data between the 5 percentile and 95 percentile

```python
# Find the 5th and 95th percentiles for each column
education=df['Education']
df['Education']=np.clip(education,np.percentile(education,5),np.percentile(education,95))
df['Education']
```

```python
age=df['Age']
df['Age']=np.clip(age,np.percentile(age,5),np.percentile(age,95))
df['Age']
```

```python
miles=df['Miles']
df['Miles']=np.clip(miles,np.percentile(miles,5),np.percentile(miles,95))
df['Miles']
```

```python
income=df['Income']
df['Income']=np.clip(income,np.percentile(income,5),np.percentile(income,95))
df['Income']
```
